```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
accident = "valentines_day_2014.csv"
vehicle = "Vehicle_Info.csv"

# Read Purchasing File and store into Pandas data frame
accident_data = pd.read_csv(accident)
vehicle_data = pd.read_csv(vehicle)
```

```python
# Create dataframe that only contains the relevant indices for merging
accident_index = accident_data[["Accident_Index"]]
accident_index
```

```python
# Merge the two files
vehicle_info = pd.merge(accident_index, vehicle_data, on="Accident_Index")
vehicle_info
```

```python
# Export to .csv file
vehicle_info.to_csv("vehicle_info_clean.csv", index = False)
```
