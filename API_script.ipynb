```python
city_url = []


for city in cities:
    url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{city}/2014-02-14/2014-02-14?unitGroup=uk&include=days&key={API_KEY}&contentType=json"

    city_url.append(url)
    
```

```python
datetime = []
address = []
lat = []
lng = []
max_temp = []
conditions = []
visibility = []
wind_speed = []
sunrise = []
sunset = []
rain = []

i = 0

for city in cities:
    
    response = requests.get(city_url[i]).json()
    
    address.append(response["address"])
    lat.append(response["latitude"])
    lng.append(response["longitude"])
    datetime.append(response["days"][0]["datetime"])
    max_temp.append(response["days"][0]["tempmax"])
    conditions.append(response["days"][0]["conditions"])
    visibility.append(response["days"][0]["visibility"])
    wind_speed.append(response["days"][0]["windspeed"])
    sunrise.append(response["days"][0]["sunrise"])
    sunset.append(response["days"][0]["sunset"])
    rain.append(response["days"][0]["precip"])
    
    i = i + 1
    
```

```python
data = {
    "City" : address,
    "Lat" : lat,
    "Lng" : lng,
    "Max Temp" : max_temp,
    "Conditions" : conditions,
    "Visibility" : visibility,
    "Wind Speed" : wind_speed,
    "Sunrise" : sunrise,
    "Sunset" : sunset,
    "Rain" : rain,
    "Date" : datetime}

data_df = pd.DataFrame(data)
data_df
```
